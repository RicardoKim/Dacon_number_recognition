```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
Root_dir = './data/'
Train_dir = Root_dir + 'train.csv'
```

```python
origin_data = pd.read_csv(Train_dir).set_index('id')
```

```python
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
# !git clone https://github.com/DHZS/tf-dropblock.git
# from tf_dropblock.nets.dropblock import DropBlock2D
```

```python
augmentation_path = Root_dir + 'new_data/'
if not os.path.exists(augmentation_path):
    os.mkdir(augmentation_path)
```

```python
shear_range = 0.2
zoom_range = 0.2
width_shift_range=0.05,
height_shift_range=0.05,
Image_generator = ImageDataGenerator(shear_range = shear_range, zoom_range = zoom_range, width_shift_range = width_shift_range, height_shift_range = height_shift_range)
```

```python
number_of_num = origin_data['digit'].value_counts().sort_index().values
max_number = np.max(number_of_num)
number_of_short = max_number - number_of_num
```

```python
number_list = []
for i in range(10):
    number_index = (origin_data['digit'] == i)
    number_list.append(origin_data[number_index].iloc[:,2:].values.reshape(-1, 28,28,1))
```

```python
for i, data in enumerate(number_list):
    save_dir = augmentation_path + '{}'.format(i) + '/'
    if not os.path.exists(save_dir):
        os.mkdir(save_dir)
    number_of_new_image = number_of_short[i] + 100
    t = tqdm(total = number_of_new_image)
    for x, val in zip(Image_generator.flow(data, save_to_dir = save_dir, save_prefix = 'aug') ,range(number_of_new_image)):
        t.update(1)
        pass
    t.close()
```
