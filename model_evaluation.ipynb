```
evaluation_model = tf.keras.models.load_model(best_model)
```

```
X_train=origin_data.iloc[:,2:].values.reshape(-1,28,28,1)
X_train = X_train/255.0
y_train = origin_data['digit'].values
```

```
def analyze_data(model, X_train, y_train):
    wrong_data = []
    for i in range(len(X_train)):
        data = X_train[i][None, :]
        predict_array = model.predict(data)
        prediction = np.argmax(predict_array)
        if (prediction != y_train[i]):
            wrong_data.append(i)
    return wrong_data
```

```
wrong_matrix = analyze_data(evaluation_model, X_train, y_train)
```

```
def analyze_data(model, X_train, y_train):
    predict_array = evaluation_model.predict(X_train)
    prediction = np.argmax(predict_array, axis = 1)
    
    error = prediction - y_train
    check_list = list(map(lambda x: x != 0, error))
    return check_list.index(True)
```
